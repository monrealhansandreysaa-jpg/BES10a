```python
#Code by Hans Andrey R. Monreal, BS Chemical Engineering 1A
#Simple 3-Day Sleep Debt Calculator
#Note: This is assuming the minimum requirement of sleep is 8 hours per day.

day1 = float(input("Enter hours slept on Day 1: "))
day2 = float(input("Enter hours slept on Day 2: "))
day3 = float(input("Enter hours slept on Day 3: "))

#Had to research about the max code for this to work properly and efficiently
sleep_debt = max(0, 24 - (day1 + day2 + day3))
print("Total sleep debt over 3 days:", sleep_debt, "hours")
```

Limit the value to Zero (Clamping)If you are performing a calculation (like health or currency) and want it to stop at 0 rather than dropping below it, you can "clamp" it.

Using Math.max (JavaScript, Python, Excel): This returns the larger of two numbers. If your result is negative, it returns 0 instead.
JavaScript/Python: value = Math.max(0, value - amount);Excel: =MAX(0, A1-B1)

In Python, max() is a built-in function used to find the largest item in an iterable or the largest of two or more arguments.
