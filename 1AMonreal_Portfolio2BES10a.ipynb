As a student who loves to play video games but wishes to balance everything from sleep and studying, I wanted to create a program that will personally help me track my hours efficiently and to give me objective feedback to help me balance my 24 hours in a day. Which is why I will be creating a tracker that will help me efficiently divide my time as a student who loves to play video games, and of course, ample sleep time.

```python
#The Program I will be creating will be set to a 24-hour day interval, so inputs over or under 24 hours will not be recognized.

# The try block checks if the user enters valid numbers
try:
    # Ask the user for study hours
    study_hours = float(input("Enter hours studied today: "))

    # Ask the user for gaming hours
    gaming_hours = float(input("Enter hours spent gaming today: "))

    # Ask the user for sleep hours
    sleep_hours = float(input("Enter hours slept today: "))

    # Idle time refers to remaining free time, such as eating, resting, commuting, scrolling, etc.

    # Calculate the total recorded hours
    total_hours = study_hours + gaming_hours + sleep_hours

    # Calculate idle time, added idle time because not all the time you are either gaming, studying, or sleeping
    idle_time = 24 - total_hours

    # Display total hours used
    print("\nTotal Recorded Hours:", total_hours)

    # Prevent impossible values above 24 hours, this ensures the user does not exceed or go under 24 hours
    if total_hours > 24:
        print("Error: Total hours cannot exceed 24 hours in one day.")

    # Continue calculations if hours are valid
    else:

        # Calculate productive hours
        total_active = study_hours + gaming_hours

        # Calculate sleep debt
        # max() prevents negative sleep debt values
        sleep_debt = max(0, 8 - sleep_hours)

        # Display the daily report
        print("\n------ DAILY REPORT ------")

        print("Study Hours:", study_hours)
        print("Gaming Hours:", gaming_hours)
        print("Sleep Hours:", sleep_hours)

        # Display calculated idle time
        print("Idle Time:", idle_time)

        print("Total Active Hours:", total_active)

        print("Sleep Debt:", sleep_debt)

        # Academic evaluation
        if study_hours >= 5:
            print("Academic Status: Very productive today!")

        elif study_hours >= 3:
            print("Academic Status: Decent study progress.")

        else:
            print("Academic Status: You should study more.")

        # Gaming evaluation
        if gaming_hours > 6:
            print("Gaming Status: Too much gaming today.")

        elif gaming_hours >= 3:
            print("Gaming Status: Balanced gaming time.")

        else:
            print("Gaming Status: Low gaming time.")

        # Sleep evaluation
        if sleep_hours >= 8:
            print("Sleep Status: Healthy sleep schedule.")

        elif sleep_hours >= 6:
            print("Sleep Status: Slight lack of sleep.")

        else:
            print("Sleep Status: Sleep deprived!")

        # Idle time evaluation
        if idle_time >= 6:
            print("Idle Status: You had a lot of free time today.")

        elif idle_time >= 3:
            print("Idle Status: Your free time was balanced.")

        else:
            print("Idle Status: Very busy day.")

        # Overall lifestyle evaluation
        if study_hours > gaming_hours and sleep_hours >= 7:
            print("Overall Result: Excellent life balance!")

        elif gaming_hours > study_hours:
            print("Overall Result: Reduce gaming and focus more on studies.")

        else:
            print("Overall Result: Try improving your daily routine.")


# If the user enters invalid input, the program will not crash but display the following message:
except:
    print("Error: Please enter numeric values only.")
```
